```python
pd.options.display.float_format = "{:.2f}".format
```

```python
df_principal = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name="Principal")

df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_ticker = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name="Ticker")
df_chatgpt = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name="ChatGPT")
```

**Aula 3 - dataframe**

```python
# Removendo colunas desnecessárias
df_principal = df_principal[['Ativo','Data','Último (R$)','Var. Dia (%)']].copy()

# Padronizando o nome das colunas
df_principal = df_principal.rename(columns={'Último (R$)':'Valor_Final', 'Var. Dia (%)': 'Var_Dia_pct'}).copy()

df_principal
```

```python
# Calculando a variação em porcentagem e o valor inicial
df_principal['Var_pct'] = df_principal['Var_Dia_pct'] / 100
df_principal['Valor_Inicial'] = df_principal['Valor_Final'] / (df_principal['Var_pct']+1)

df_principal
```

```python
# Buscando informações da tabela 'Total_de_ações'
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')

# Removendo coluna desnecessária, padronizando o nome e o tipo da coluna 'Qtde, Teórica'
df_principal= df_principal.drop(columns=['Código'])
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtde_Teorica'}).copy()
df_principal['Qtde_Teorica'] = df_principal['Qtde_Teorica'].astype(int)

df_principal
```

```python
# Calculando a variação em reais
df_principal['Variacao_RS'] = (df_principal['Valor_Final'] - df_principal['Valor_Inicial']) * df_principal['Qtde_Teorica']

df_principal
```

```python
# Analisando se as ações apresentaram lucro ou prejuízo
df_principal['Resultado'] = df_principal['Variacao_RS'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estagnou'))

df_principal
```

```python
# Buscando informações da tabela 'Ticker' e removendo colunas desnecessária
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns={'Ticker'})

df_principal
```

```python
# Buscando informações da tabela 'ChatGPT', renomeando colunas e removendo desnecessária
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Empresa', how='left')
df_principal = df_principal.drop(columns={'Empresa'})
df_principal = df_principal.rename(columns={'Idade (em anos)':'Idade'}).copy()

df_principal
```

```python
# Analisando a categoria de idade das empresas
df_principal['Cat_Idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100 anos' if x > 100 else ('Entre 50 e 100 anos' if x > 50 else 'Menos de 50 anos'))

df_principal
```

**Análises**

```python
# Calculando o maior valor da coluna 'Variacao_RS'
maior = df_principal['Variacao_RS'].max()

# Calculando o menor valor da coluna 'Variacao_RS'
menor = df_principal['Variacao_RS'].mean()

# Calculando a média da coluna 'Variacao_RS'
media = df_principal['Variacao_RS'].mean()

# Calculando a média dos valores da coluna 'Variacao_RS' onde 'Resultado' é 'Subiu'
media_subiu = df_principal.loc[df_principal['Resultado'] == 'Subiu', 'Variacao_RS'].mean()

# Calculando a média dos valores da coluna 'Variacao_RS' onde 'Resultado' é 'Desceu'
media_desceu = df_principal.loc[df_principal['Resultado'] == 'Desceu', 'Variacao_RS'].mean()

print('Maior\tR$ {:.2f}'.format(maior))
print('Menor\tR$ {:.2f}'.format(menor))
print('Média\tR$ {:.2f}\t'.format(media))
print('Média que subiu\tR$ {:.2f}\t'.format(media_subiu))
print('Média que desceu\tR$ {:.2f}\t'.format(media_desceu))
```

```python
# Cria um dataframe com apenas as empresas que apresentaram lucro
df_principal_subiu = df_principal.loc[df_principal['Resultado'] == 'Subiu']

df_principal_subiu
```

```python
# Agrupa os dados do dataframe anterior pelo segmento das empresas
df_analise_segmentos = df_principal_subiu.groupby('Segmento')['Variacao_RS'].sum().reset_index()

df_analise_segmentos
```

```python
# Agrupa os dados do dataframe original pelo resultado da empresa
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_RS'].sum().reset_index()
df_analise_saldo
```

**Gráficos**

```python
# Cria um gráfico de barras da variação em reais pelo resultado positivo ou negativo das empresas
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_RS', text='Variacao_RS', title='Variação em reais por resultado - Barra')
# Formata monetariamente o template correspondente à variação de cada gráfico
fig.update_traces(texttemplate='R$ %{y:,.2f}', textposition='outside')

fig.show()
```

```python
# Cria e exibe um gráfico de pizza com a variação em reais dos segmentos das empresas
fig2 = px.pie(df_analise_segmentos, values='Variacao_RS', names='Segmento', title='Variação em reais por segmento - Pizza')

fig2.show()
```
